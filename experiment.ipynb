```python
import requests
import geopandas as gpd

# Overpass API endpoint
url = 'http://overpass-api.de/api/interpreter'

# Overpass query for shops in Tbilisi
query = """
[out:json];
node["shop"](41.5,44.7,41.8,45.0);  # Bounding box for Tbilisi
out body;
"""

# Send request to Overpass API
response = requests.get(url, params={'data': query})
data = response.json()

# Convert to GeoDataFrame
nodes = data['elements']
shops = []
for node in nodes:
    shops.append({
        'name': node.get('tags', {}).get('name', 'Unknown'),
        'lat': node['lat'],
        'lon': node['lon']
    })

# Create DataFrame
shops_df = pd.DataFrame(shops)
print(shops_df)
```
